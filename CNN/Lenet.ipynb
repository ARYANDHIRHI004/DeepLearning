```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Sequential
```

```python
(X_train, y_train), (X_test,y_test) = mnist.load_data()
```

```python
X_trian = X_train/255
X_test = X_test/255
```

```python
from tensorflow.keras.utils import to_categorical

num_classes = 10
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
model = Sequential()

model.add(Conv2D(6, kernel_size=(5,5), padding='same' ,activation='tanh',input_shape = (28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(6, kernel_size=(5,5), padding='same' ,activation='tanh'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(120, activation='tanh'))
model.add(Dense(84, activation='tanh'))
model.add(Dense(10, activation='softmax'))


model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'] )
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

# train the model
checkpointer = ModelCheckpoint(filepath='model.weights.best.keras', verbose=1,
                               save_best_only=True)
hist = model.fit(X_train, y_train, batch_size=64, epochs=10,
          validation_data=(X_test, y_test), callbacks=[checkpointer],
          verbose=2, shuffle=True)
```
